```python
import numpy
from numpy import random
import pandas
from keras.models import Sequential, load_model
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
# load dataset
dataframe = pandas.read_csv("iris.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]

testDataFrame = pandas.read_csv("test-iris.csv", header=None)
test_X = dataset[:,0:4].astype(float)
test_Y = dataset[:,4]
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
```

```python
print("Name:", Y[100])
print("Encoded:", encoded_Y[100])
print("One-Hot Encoded:", dummy_y[100], "\n")

print("Name:", Y[1])
print("Encoded:", encoded_Y[1])
print("One-Hot Encoded:", dummy_y[1], "\n")

print("Name:", Y[50])
print("Encoded:", encoded_Y[50])
print("One-Hot Encoded:", dummy_y[50], "\n")
```

```python
#create model
model = Sequential()
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, numpy.array(dummy_y), validation_split=0.2, epochs=200, shuffle=True)
```

```python
model.save('iris.h5')
del model
```
